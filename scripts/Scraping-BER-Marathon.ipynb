# Scraping BMW Berlin Marathon

```python
import pandas as pd
import requests
import time
```

```python
url = "https://api.results.scc-events.com/result"

params = {
    "ek": "BM",
    "ci": "BML",
    "y": 2024,
    "t": "BM_2024",
    "order[0][column]": "0",
    "order[0][dir]": "asc",
    "order[0][name]": "",
    "start": 0,
    "length": 20,
    "search[value]": "",
    "search[regex]": "false",
}

response = requests.get(url, params = params)
```

```python
output = response.json()
```

```python
len(output["data"])
```

```python
output
```

```python
data_list= []
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2021, 2022, 2023, 2024]

for year in years:
    print(f'Scrape {year} …')
    if year <= 2019:
        params['ci'] = "MAL"
    else:
        params['ci'] = "BML"  
    params['y'] = year
    params['t'] = "BM_" + str(year)

    params['start'] = 0 

    for i in range(5000):
        if i % 50 == 0:
            print(f'--- {i}')
        response = requests.get(url, params = params)
        params['start'] += 20
        output = response.json()
        output_data = output['data']
        # Gibt es noch Elemente in output['data']?
        if len(output_data) == 0:
            break
        for d in output_data:
            d['year'] = year
        data_list += output_data
        time.sleep(0.05)
```

```python
df = pd.DataFrame(data_list)
```
